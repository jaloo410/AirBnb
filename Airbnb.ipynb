```python
import pandas as pd
listings = pd.read_csv("Listings.csv", encoding = "ISO-8859-1", low_memory= False, parse_dates= ["host_since"])
```

```python
listings.head()
```

```python
listings.info()
```

```python
paris_listings = listings.query("city == 'Paris'").loc[:,["host_since", "neighbourhood", "city", "accommodates", "price"]]
paris_listings.info()
```

```python
paris_listings.isna().sum()
```

```python
paris_listings.describe()
```

```python
paris_listings.query("price == 0 and accommodates == 0").count()
```

```python
paris_listings_neighbourhood = (paris_listings.groupby("neighbourhood").agg({"price": "mean"})).sort_values("price")
paris_listings_neighbourhood.tail()
```

```python
paris_listings_accommodates = (paris_listings.query("neighbourhood == 'Elysee'").groupby("accommodates").agg({"price":"mean"}).sort_values("price"))
paris_listings_accommodates.tail()
```

```python
paris_listings_over_timetime = (paris_listings.set_index("host_since").resample("YE")).agg({"neighbourhood": "count", "price" : "mean"})
paris_listings_over_timetime.head()
```

```python
import seaborn as sns
(paris_listings_neighbourhood
.plot
.barh(
    title="Average Listing Price by Paris Neighbourhood",
    xlabel="Price per Night (Euros)",
    ylabel="Neighbourhood",
    legend=None
)
    )
sns.despine()
```

```python
(paris_listings_accommodates
.plot
.barh(
    title="Average Listing Price by Paris Accommodates",
    xlabel="Price per Night (Euros)",
    ylabel="Accommodation Capacity",
    legend=None
)
    )
sns.despine()
```

```python
(paris_listings_over_timetime["neighbourhood"].plot(ylabel = "New Hosts", title = "New Airbnb Hosts in Paris Over Time"))
sns.despine()
```

```python
(paris_listings_over_timetime["price"].plot(ylabel = "Average Price (Euros)", title = "Average AirBnb Price in Paris Over Time"))
sns.despine()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(paris_listings_over_timetime.index, paris_listings_over_timetime["neighbourhood"], label = "New Hosts", c = "red")
ax.set_ylabel("New Hosts")

ax2 = ax.twinx()

ax2.plot(paris_listings_over_timetime.index, paris_listings_over_timetime["price"], label="Average Price",c = "yellow")
ax2.set_ylim(0)
ax2.set_ylabel("Average Price")
ax.set_title("2015 Regulations Lead to Fewer New Hosts, Higher Prices")
```
